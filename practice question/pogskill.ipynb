```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = [0,0]
c = [[1,0],[0,1]]
x,y = np.random.multivariate_normal(m,c,100).T
plt.plot(x,y,'.')
plt.axis('equal')
plt.show()
```

```python
m = [-1,1]
c = [[1,0],[0,1]]
x,y = np.random.multivariate_normal(m,c,100).T
plt.plot(x,y,'.')
plt.axis('equal')
plt.show()
#shift toward (-1,1)
```

```python
m = [0,0]
c = [[2,0],[0,2]]
x,y = np.random.multivariate_normal(m,c,100).T
plt.plot(x,y,'.')
plt.axis('equal')
plt.show()
#more concentrated
```

```python
m = [0,0]
c = [[1,0.5],[0.5,1]]
x,y = np.random.multivariate_normal(m,c,100).T
plt.plot(x,y,'.')
plt.axis('equal')
plt.show()
#points along x=y
```

```python
m = [0,0]
c = [[1,-0.5],[-0.5,1]]
x,y = np.random.multivariate_normal(m,c,100).T
plt.plot(x,y,'.')
plt.axis('equal')
plt.show()
#points along -x=y
```
